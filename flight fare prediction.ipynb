```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
stdsc=StandardScaler()
X_train=stdsc.fit_transform(X_train)
X_test=stdsc.transform(X_test)
```

```python
y_test.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_mod=RandomForestRegressor()
rf_mod.fit(X_train,y_train)
y_pred=rf_mod.predict(X_test)
print(y_pred)
y_pred.mean()
```

```python
rf_mod.score(X_test,y_test)
```

```python
rf_mod.score(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
cores = cross_val_score(rf_mod,X,y,scoring="neg_mean_squared_error", cv=10)
```

```python
rmse_cores=np.sqrt(-cores)
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
```

```python
display_scores(rmse_cores)
```

```python
pores = cross_val_score(dec_mod,X,y,scoring="neg_mean_squared_error", cv=10)
rmse_pores=np.sqrt(-pores)
display_scores(rmse_pores)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = [
    {'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]},
    {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]},
  ]


grid_search = GridSearchCV(rf_mod, param_grid,
                           scoring='neg_mean_squared_error',
                           return_train_score=True,
                           cv=10,
                          )

grid_search.fit(X,y)
```

```python
grid_search.best_params_
```

```python
cv_scores = grid_search.cv_results_

##printing all the parameters along with their scores
for mean_score, params in zip(cv_scores['mean_test_score'], cv_scores["params"]):
    print(np.sqrt(-mean_score), params)
```
